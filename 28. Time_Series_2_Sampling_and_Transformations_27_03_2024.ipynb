#### Upsampling Data

```python
# upsample to daily intervals
from pandas import read_csv
# from pandas import datetime
import datetime
from matplotlib import pyplot
```

```python
series = read_csv('sales (1).csv', header=0, index_col=0, parse_dates=True)  #,squeeze=True
```

```python
series
```

```python
# We have month wise data, we need to break it down into day wise
# Resample data day wise and use linear interpolation - that is mean wise
# On an average, we split the data day wise linearly
upsampled = series.resample('D').mean()
print(upsampled.head(32))
```

##### interpolate the missing value

```python
# Interpolating the data linearly using mean
interpolated = upsampled.interpolate(method='linear')
print(interpolated.head(32))
interpolated.plot()
pyplot.show()
```

#### Downsampling Data

```python
# downsample to quarterly intervals
from pandas import read_csv
import datetime
from matplotlib import pyplot
```

```python
# Resampling the data quarterwise and we are taking means for every quarter
resample = series.resample('Q')
quarterly_mean_sales = resample.mean()
quarterly_mean_sales
```

```python
print(quarterly_mean_sales.head())
quarterly_mean_sales.plot()
pyplot.show()
```

# Tranformations

```python
# load and plot a time series
from pandas import read_csv
from matplotlib import pyplot
```

```python
series = read_csv('airline-passengers.csv', header=0, index_col=0, parse_dates=True)
```

```python
# line plot
# using the subplot
pyplot.subplot(211) # 211 means 2 rows, 1 column, 1st visualizations
pyplot.plot(series)
# histogram
pyplot.subplot(212) # 212 means 2 rows, 1 column, 2nd visualizaton
pyplot.hist(series)
pyplot.show()
```

#### Square Root Transform

```python
from pandas import read_csv
from pandas import DataFrame
from numpy import sqrt
from matplotlib import pyplot
```

```python
# Applying the square root transformation on the data using the numpy library
dataframe = DataFrame(series.values)
dataframe.columns = ['passengers']
dataframe['passengers'] = sqrt(dataframe['passengers'])
```

```python
# Plotting the square root transformed data
# line plot
pyplot.subplot(211)
pyplot.plot(dataframe['passengers'])
# histogram
pyplot.subplot(212)
pyplot.hist(dataframe['passengers'])
pyplot.show()
```

#### Log Transform

```python
# Applying log transformation on the data
from numpy import log
dataframe = DataFrame(series.values)
dataframe.columns = ['passengers']
dataframe['passengers'] = log(dataframe['passengers'])

# Plotting the log transformed data
# line plot
pyplot.subplot(211)
pyplot.plot(dataframe['passengers'])
# histogram
pyplot.subplot(212)
pyplot.hist(dataframe['passengers'])
pyplot.show()
```

```python
# This transformation is better as we are getting a more normally distributed data
```
