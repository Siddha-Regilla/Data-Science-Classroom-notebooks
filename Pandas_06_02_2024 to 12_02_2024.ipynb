```python
#Type 4
df1.iloc[0:10,[0,2,5]] #1st 10 records (starting from 0th index uptil 9th index) and only the specified number index columns
```

How to handle duplicates

```python
#Identify duplicates
df2.duplicated()
```

```python
#Checking those duplicated records
df2[df2.duplicated()]
```

```python
#Dropping the duplicates
df3=df2.drop_duplicates()
```

```python
df3
```

```python
#Rearranging the dataframe to reflect the updated indexes
df4=df3.reset_index()
```

```python
df4
```

```python
#Removing the index column showing the old indexes
df5=df4.drop('index',axis=1)
```
