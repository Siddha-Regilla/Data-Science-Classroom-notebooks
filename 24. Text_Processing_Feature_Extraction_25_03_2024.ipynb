```python
# Upload dataset
import pandas
book = pd.read_csv("apple.txt",on_bad_lines = 'skip')
book
```

```python
# Pre-processing
# Cleaning of the data
# strip function helps remove unwanted characters like emply strings, spaces
book = [x.strip() for x in book.x] # remove both the leading and the trailing characters
book = [x for x in book if x] # removes empty strings, because they are considered in Python as False
book[0:10]
```

```python
# All the reviews are in seperate quotation marks. But we want to put them together in a list
# Joining the list into one string/text
text = ' '.join(book)
text
```

```python
# Removing Punctuation marks. All the punctuation marks are replaced with spaces
no_punc_text = text.translate(str.maketrans('', '', string.punctuation)) #with arguments (x, y, z) where 'x' and 'y'
# must be equal-length strings and characters in 'x'
# are replaced by characters in 'y'. 'z'
# is a string (string.punctuation here)
no_punc_text
```

```python
# Tokenization function from nltk library.
# Breaking the sentences into tokens
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize
text_tokens = word_tokenize(no_punc_text)
print(text_tokens[0:50])
```

```python
# Total length of tokens
len(text_tokens)
```

```python
# Remove stopwords
import nltk
from nltk.corpus import stopwords
nltk.download('punkt')
nltk.download('stopwords')

my_stop_words = stopwords.words('english') # mention the language in the brackets
my_stop_words.append('the') # Mention any additional stop words that you want to removed from the corpus in the brackets
no_stop_tokens = [word for word in text_tokens if not word in my_stop_words] # saving the non stop words in a new variable
print(no_stop_tokens[0:40])
```

```python
# Noramalize the data
# Converting them into lower case
lower_words = [x.lower() for x in no_stop_tokens]
print(lower_words[0:25])
```

```python
# Stemming
# Chopping off the ends of the words. In stemming it gets it to a meaningless root form.
from nltk.stem import PorterStemmer
ps = PorterStemmer()
stemmed_tokens = [ps.stem(word) for word in lower_words]
print(stemmed_tokens[0:40])
```

```python
# Instead of chopping off using stemming, where the base word has no real meaning, we can apply lemmatization
# Loading the language model
# NLP english language model of spacy library
nlp = spacy.load('en_core_web_sm')
```

```python
# Applying the language model on the tokens
# lemmas being one of them, but mostly POS, which will follow later
doc = nlp(' '.join(no_stop_tokens))
print(doc[0:40])
```

```python
# Lemmatizing the words
lemmas = [token.lemma_ for token in doc]
print(lemmas[0:25])
```

#### Feature Extraction

```python
# Apply count vectorizers on the lemmas
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(lemmas)
```

```python
# the no. of times each tokeen has been used
print(vectorizer.vocabulary_)
```

```python
# Feature extraction matrix with all the lemmas
print(vectorizer.get_feature_names_out()[50:100])
print(X.toarray()[50:100])
```

```python
# Feature extraction matrix size
print(X.toarray().shape)
```

#### Let's see how can bigrams and trigrams can be included here

```python
# Applying feature extraction on text data (rather than lemmas)
vectorizer_ngram_range = CountVectorizer(analyzer='word',ngram_range=(1,3),max_features = 100)
bow_matrix_ngram =vectorizer_ngram_range.fit_transform(book)
```

```python
# Printing the feature extraction data
print(vectorizer_ngram_range.get_feature_names_out())
print(bow_matrix_ngram.toarray())
```

```python
# The above matrix can be used in the machine learning model for classification problem
```

#### TFidf vectorizer

```python
# Similar to countvectorizer -
# Apply TF-IDF vectorizer, specify regularization, it analyses 'word', give a ngram_range and the max no. of features
# Apply it on to our data
# Get the features
# Convert it into a matrix
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer_n_gram_max_features = TfidfVectorizer(norm="l2",analyzer='word', ngram_range=(1,3), max_features = 500)
tf_idf_matrix_n_gram_max_features =vectorizer_n_gram_max_features.fit_transform(book)
print(vectorizer_n_gram_max_features.get_feature_names_out())
print(tf_idf_matrix_n_gram_max_features.toarray())
```

####  Generate wordcloud

```python
# Creating a word cloud
# This is to visualise the words used in the review
# Import packages
import matplotlib.pyplot as plt
%matplotlib inline
from wordcloud import WordCloud, STOPWORDS
# Define a function to plot word cloud
def plot_cloud(wordcloud):
    # Set figure size
    plt.figure(figsize=(40, 30))
    # Display image
    plt.imshow(wordcloud)
    # No axis details
    plt.axis("off");
```

```python
# Generate wordcloud
# Higher the size of the word, the more no. of times people are talking about
stopwords = STOPWORDS
stopwords.add('will')
wordcloud = WordCloud(width = 3000, height = 2000, background_color='black', max_words=100,colormap='Set2',stopwords=stopwords).generate(text)
# Plot
plot_cloud(wordcloud)
```
