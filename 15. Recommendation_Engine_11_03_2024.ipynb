```python
# Importing Libraries
import pandas as pd
import numpy as np
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
#from scipy.spatial.distance import cosine, correlation
```

# ***User based Colaborative filtering***

```python
# Upload dataset - Movie (1).csv
# Read dataset
movies_df = pd.read_csv('Movie (1).csv')
movies_df = movies_df.sort_values(by ='userId').reset_index(drop = True) # Sorting the values based on userID and re arranging the index
movies_df
```

```python
movies_df[0:5]
```

```python
#number of unique users in the dataset
len(movies_df.userId.unique())
```

```python
# Number of unique movies
len(movies_df.movie.unique())
```

```python
# Creating the User-Item interaction matrix
# We are using the pivot function and saying that the userID should be the index, Each individual movie should be the column and rating should be the values
user_movies_df = movies_df.pivot(index='userId',
                                 columns='movie',
                                 values='rating').reset_index()
```

```python
user_movies_df
```

```python
# As the index values are randomised we are changing them to unique userID values
user_movies_df.index = movies_df.userId.unique()
```

```python
user_movies_df
```

```python
#Impute those NaNs with 0 values
user_movies_df.fillna(0, inplace=True)
```

```python
user_movies_df
```

```python
# We are using pairwise to find out cosine similarity
# Pairwise distances give the dismilarity between the obs. Therefore, we need to do 1 - pairwise to help calculate the similarity
user_sim = 1 - pairwise_distances( user_movies_df.values,metric='cosine')
```

```python
# From the above code we get the user similarity matrix
user_sim
```

```python
#Store the matrix results in a dataframe
user_sim_df = pd.DataFrame(user_sim)
```

```python
#Set the index (row names) and column names to user ids
user_sim_df.index = movies_df.userId.unique()
user_sim_df.columns = movies_df.userId.unique()
```

```python
# All the values in the diagonal are 1 as they are similar
user_sim_df.iloc[0:5, 0:5]
```

```python
# As the diagonal values are irrelavent for us, we are replacing them with 0
np.fill_diagonal(user_sim, 0)
user_sim_df.iloc[0:5, 0:5]
```

```python
# Now we are trying to identify which are the Most Similar Users
# To do that we are comparing each row userID with the maximum cosine value userID to find out similar users
user_sim_df.idxmax(axis=1)[0:5]
```

```python
# From the above table we have identified some similar users
# We are now comparing 1 pair of similar users - 4 & 8
movies_df[(movies_df['userId']==4) | (movies_df['userId']==8)]
```

```python
# So we can see that both of them have watched 2 movies. '8' has watched 2 more than '4'.
# So we can recommend the other 2 movies that '4' hasn't watched based on their similarity
```

```python
user_1=movies_df[movies_df['userId']==4]
```

```python
user_2=movies_df[movies_df['userId']==8]
```

```python
user_2.movie
```

```python
user_1.movie
```

```python
pd.merge(user_1,user_2,on='movie',how='outer')
```

# ***Item Based Colaborative Filtering***

```python
#IBCF
#Fetching ratings for Toy Story
sabrina_user_rating = user_movies_df['Sabrina (1995)']
```

```python
sabrina_user_rating
```

```python
#Finding the correlation with different movies and Sabrina
similar_to_sabrina = user_movies_df.corrwith(sabrina_user_rating)
```

```python
# Other movies that have some form of correlation with Sabrine
similar_to_sabrina.head(10)
```

```python
# We put the above data in a dataframe
corr_sabrina = pd.DataFrame(similar_to_sabrina, columns=['Correlation'])
corr_sabrina.dropna(inplace=True)
corr_sabrina.head()
```

```python
# And we focus on those movies that have positive correlation
corr_sabrina[corr_sabrina['Correlation'] > 0].sort_values(by='Correlation', ascending=False).head(5)
```
