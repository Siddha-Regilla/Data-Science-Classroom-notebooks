```python
#Importing Library
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA #To perform PCA
import matplotlib.pyplot as plt # for visualizations
import seaborn as sns # For visualizations
from sklearn.preprocessing import scale # Similar to standard scaler
```

***PCA is a Linear dimension reduction technique***

upload the dataset - Universities_Clustering



```python
# Uploading the dataset - Universities_Clustering.csv
# Read the dataset
uni = pd.read_csv("Universities_Clustering.csv")
uni.describe()
uni.head()
```

```python
# Considering only numerical data
uni.data = uni.iloc[:,1:] # selecting the 6 numeric columns from index 1
uni.data.head()
# Converting into numpy array
UNI = uni.data.values
UNI
```

```python
# Normalizing / standardizing the numerical data
uni_normal = scale(UNI)
uni_normal
```

```python
# Using the PCA function on the standardized data
pca = PCA(n_components = 6) # no. of PC to be created needs to be mentioned
pca_values = pca.fit_transform(uni_normal) # fitting and transforming the standardized data
pca_values
# converting the data into PC
```

```python
# To find out the loadings or weights used in the PC
pca.components_
```

```python
# The amount of variance or the amount of information that each PCA explains is
var = pca.explained_variance_ratio_
var
```

```python
# Converting the values in to percentages and then finding out the Cumulative variance
var1 = np.cumsum(np.round(var,decimals = 4)*100)
var1
# The output is an array as follows - PC1 , PC1+PC2, PC1+PC2+PC3.... an dso on
```

```python
# Variance plot for PCA components obtained
plt.plot(var1,color="red")
```

```python
 # Selecting the 1st 2 PCs and dropping the rest
 pca_values[:,0:2]
```

```python
# New dataset - from 6 dimensions we have compressed the data to 2 dimensions
finalDf = pd.concat([pd.DataFrame(pca_values[:,0:2],columns=['pc1','pc2']), uni[['Univ']]], axis = 1)
finalDf
```

```python
plt.style.use('classic')
```

```python
# Using the new dataset to plot a scatter plot
sns.scatterplot(data=finalDf,x='pc1',y='pc2',hue='Univ',s = 100)
```

```python
# The above observation is not clear because of the colours. So we created another code with university names
p1 = sns.scatterplot(data=finalDf,x='pc1',y='pc2',s = 100)
for line in range(0,finalDf.shape[0]):
     p1.text(finalDf.pc1[line], finalDf.pc2[line], finalDf.Univ[line], horizontalalignment='left', size='medium')
```

# The above PCA analysis, following are the insights -

1. We can roughly understand the grouping of data if we are to apply the clustering model onto the data. Or basically it helps us spot the patterns in our data

2. After performing PCA, we can use the dataset to develop any model. Be it - linear regression, clustering, etc

3. As this is a kind of transformation of data, it can only be applied on numeric columns. This can't be applied on categorical data
